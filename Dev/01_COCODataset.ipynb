```python
# random pick one image and show with cor and labels
idx = np.random.randint(0,len(d['images']))  #random pick
import os
image = os.path.join(img_path, d['images'][idx]['file_name'])  #image path
img = mpimg.imread(image) #read image

image_idx = d['images'][idx]['id'];image_idx #fet image idx
#get bbx and corl
corl = []
categories = []
for l in d['annotations']:
    if int(l['image_id']) == image_idx:
        corl.append(l['bbox'])
        categories.append(l['category_id'])
        #print(l['bbox'],l['category_id'])
        
Cat = []
for cc in categories:
    for an in d['categories']:
        if cc == an['id']:
            Cat.append(an['name'])

font = cv2.FONT_HERSHEY_SIMPLEX        
for cor, cat in zip(corl,Cat):
    x, y, w, h = cor
    cv2.rectangle(img, (int(x),int(y)), (int(x+w),int(y+h)), (255,0,0), 2) # image, 左上顶点, 右下顶点, color, line_length
    cv2.putText(img, cat, (int(x),int(y)), font, 1, (255, 255, 0), 1, cv2.LINE_AA)
plt.imshow(img)
#plt.axis('off')
plt.show()
```
